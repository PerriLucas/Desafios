```python
import pandas as pd
```

```python
categorias = pd.read_csv('categories.csv', sep = ';', index_col=('category_id'))
categorias1 = categorias.drop(columns = 'picture')
categorias1
```

```python
customers = pd.read_csv('customers.csv', sep = ';', index_col=('customer_id'))
customers['region'] = customers['region'].fillna('N/A')
customers['fax'] = customers['fax'].fillna('N/A')
customers['postal_code'] = customers['postal_code'].fillna('N/A')
customers
```

```python
emp_ter = pd.read_csv('employee_territories.csv', sep = ';',index_col = 'employee_id')
emp_ter
```

```python
emp = pd.read_csv('employees.csv', sep=';', index_col=('employee_id'))
emp1 = emp.drop(columns = 'photo')
emp1['region'] = emp1['region'].fillna('N/A')
emp1['reports_to'] = emp1['reports_to'].fillna('N/A')
emp1
```

```python
order_details = pd.read_csv('order_details.csv', sep=';', index_col='order_id')
order_details
```

```python
orders = pd.read_csv('orders.csv', sep=';', index_col=('order_id'))
orders['ship_region'] = orders['ship_region'].fillna('N/A')
orders['shipped_date'] = orders['shipped_date'].fillna('Not yet shipped')
orders['ship_postal_code'] = orders['ship_postal_code'].fillna('N/A')
orders
```

```python
prod = pd.read_csv('products.csv', sep=';', index_col = 'product_id')
prod
```

```python
region = pd.read_csv('region.csv', sep =';', index_col='region_id')
region
```

```python
shippers = pd.read_csv('shippers.csv', sep=';', index_col='shipper_id')
shippers
```

```python
suppliers = pd.read_csv('suppliers.csv', sep=';',index_col=('supplier_id'))
suppliers['region']=suppliers['region'].fillna('N/A')
suppliers['fax']=suppliers['fax'].fillna('N/A')
suppliers = suppliers.drop(columns = 'homepage')
suppliers
```

```python
ter = pd.read_csv('territories.csv', sep = ';', index_col=('territory_id'))
ter
```

```python
us_states = pd.read_csv('us_states.csv', sep=';',index_col=('state_id'))
us_states
```

```python
with pd.ExcelWriter('Northwind.xlsx') as writer: 
  categorias1.to_excel(writer,
             sheet_name='Categorias') 
  customers.to_excel(writer,
             sheet_name='Customers')  
  emp_ter.to_excel(writer,
             sheet_name='Emp_Territories') 
  emp1.to_excel(writer,
             sheet_name='Employees') 
  order_details.to_excel(writer,
             sheet_name='Order Details')
  orders.to_excel(writer,
             sheet_name='Orders')  
  prod.to_excel(writer,
             sheet_name='Products')
  region.to_excel(writer,
             sheet_name='Region') 
  shippers.to_excel(writer,
             sheet_name='Shippers')  
  suppliers.to_excel(writer,
             sheet_name='Suppliers') 
  ter.to_excel(writer,
             sheet_name='Territories') 
  us_states.to_excel(writer,
                   sheet_name='US States')
```
